# 2023 Dataset 
* School Dataset obtained from: https://discover.data.vic.gov.au/dataset/school-locations-2023

```python
import pandas as pd
import matplotlib as plt
```

```python
schools_df = pd.read_csv('../data/landing/dv346-schoollocations2023.csv', encoding='cp1252')
schools_df
```

```python
schools_df.columns
```

```python
# remove redundant features not useful for our analysis
schools_df = schools_df.drop(['Education_Sector', 'Entity_Type', 'School_No', 'School_Type', 'School_Status', 
                            'Address_Line_2', 'Address_State', 'Postal_Address_Line_1', 'Postal_Address_Line_2', 
                            'Postal_Town', 'Postal_State', 'Postal_Postcode', 'Full_Phone_No'], axis=1)
schools_df
```

```python
# save our preprocessed schools to `curated` folder
schools_df.to_csv('../data/curated/preprocessed_school.csv', index=False)
```

```python
# read our cleaned schools while ignoring the 
schools = pd.read_csv('../data/curated/preprocessed_school.csv', encoding='cp1252')
schools
```

```python
# total schools by postcode in Victoria
schools.groupby('Address_Postcode')['School_Name'].count()
```

```python
# total schools by suburbs in Victoria
schools.groupby('Address_Town')['School_Name'].count()
```
